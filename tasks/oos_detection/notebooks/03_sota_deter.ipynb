# SOTA: DETER

## Статус: отложено

DETER (Zawbaa et al., LREC-COLING 2024) запланирован как
SOTA-модель (верхняя граница качества).

Интеграция отложена до запуска на full train из-за
сложности зависимостей репозитория (USE + TSDAE).

**Ссылки:**
- Paper: [arxiv.org/abs/2405.19967](https://arxiv.org/abs/2405.19967)
- GitHub: [Hossam-Mohammed-tech/Intent_Classification_OOS](https://github.com/Hossam-Mohammed-tech/Intent_Classification_OOS)

**Планируемые метрики на CLINC150** (из статьи):
- F1 known intents: +13% vs ADB
- F1 unknown (OOS): +5% vs ADB

DETER будет добавлен в таблицу результатов при запуске
full train эксперимента.
